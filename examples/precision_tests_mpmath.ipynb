### Understanding the mpmath package

[Scott: start with an introduction before the code. Explain the purpose of this notebook.]

The [mpmath](https://mpmath.org/) is a Python library providing arbitrary-precision floating-point arithmetic for real and complex numbers (by [Fredrik Johansson](https://fredrikj.net/)). It can be integrated into Python projects directly or used via [SymPy](https://www.sympy.org/en/index.html) or [Sage](https://www.sagemath.org/). To achieve arbitrarily precise calculations of the quantum harmonic oscillator wave function, an arbitrary-precision module was integrated into the Fast Wave package using mpmath.

```python
#@title  Fast Wave: Single Fock & Single Position to a Complex Output & Complex Input (arbitrary precision)
n = 10 #@param
real_max = 5.0 #@param
real_min = -5.0 #@param
imag_max = 5.0 #@param
imag_min = -5.0 #@param
number_of_points = 2000 #@param
prec = 1000 #@param

plt.figure(figsize=(10, 10))

fig = plt.figure()
fig.set_size_inches(8, 6)

cplot(lambda z: wm.psi_n_single_fock_single_position_complex(n,z,prec=prec), re = [real_min, real_max],
      im = [imag_min, imag_max], points = number_of_points)

```
